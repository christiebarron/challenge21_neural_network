## Preprocessing

```python
# Import our dependencies

import pandas as pd
import numpy as np

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
import tensorflow as tf
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.columns = map(str.lower, application_df.columns)
application_df = application_df.drop(columns = ["ein", "name"])
application_df.head(10)
```

```python
# Determine the number of unique values in each column.
display(application_df.nunique())
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df["application_type"]
application_df.groupby("application_type").count()
application_df["application_type"].value_counts()
```

```python
#create list of variables to keep 
to_keep = ["T3", "T4", "T6", "T5", "T19", "T8", "T7", "T10"]

#create list of variables to replace
application_types_to_replace = ["T17", "T15", "T29", "T14", "T25", "T2", "T12", "T13", "T9"]
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#application_types_to_replace = application_df["application_type"].unique()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['application_type'] = application_df['application_type'].replace(app,"Other")

# Check to make sure binning was successful
print(application_df['application_type'].value_counts())

```

```python
#change pandas settings to print all rows to make seeing the value counts easier
pd.set_option('display.max_rows', None)

#check value counts of the classification variable
application_df[["classification"]].value_counts()
```

```python
#create a list of values in classification to keep
to_keep = ["C1000", "C2000", "C1200", "C3000", "C2100", "C7000"]

#create a list of all unique values in 
classificaiton_vars = application_df["classification"].unique()
```

```python
#use numpy to create a list of "to replace" values not in the "to_keep" list
classifications_to_replace = np.setdiff1d(classificaiton_vars,to_keep)
classifications_to_replace
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['classification'] = application_df['classification'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['classification'].value_counts()
```
